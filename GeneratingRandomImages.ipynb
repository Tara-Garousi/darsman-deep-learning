## Generating random three-channel images

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_number = 5
channel_number = 3
image_width = 28
image_height = 28
input_shape = (sample_number, image_width, image_height, channel_number)
input_data = tf.random.normal(input_shape)
print(input_data.shape)
print(input_data[0])
```

```python
rgb_image = np.zeros((image_width, image_height, 3))
print(rgb_image.shape)
print(rgb_image)
```

```python
for j in range(sample_number):
  rgb_image[:, :, :] = 0
  colors = ['Red', 'Green', 'Blue']
  for channel_index in range(channel_number):
      data = input_data[0, :, :, channel_index].numpy()
      data = np.clip(data, 0, 1)                            # Clip the data and convert it to numbers in the range 0 to 1
      # rgb_image[:, :, :] = 0                                # Zero all channels
      rgb_image[:, :, channel_index] = data                 # Setting the desired channel

      plt.imshow(rgb_image)                                 # Grayscale image display with Matplot library
      plt.title(f"Input Channel ({colors[channel_index]})")
      plt.axis('off')
      plt.show()
```
